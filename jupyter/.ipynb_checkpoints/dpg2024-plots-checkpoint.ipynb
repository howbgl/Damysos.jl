```python
def find_files(directory, filename):
    """
    Recursively searches a directory and all its subdirectories for a file with a given name.
    Returns a list of all matching file paths.
    """
    matches = []
    
    # Walk through the directory and its subdirectories
    for root, dirs, files in os.walk(directory):
        for file in files:
            if file == filename:
                matches.append(os.path.join(root, file))
    
    return matches
```

# Schematic wide & narrow bands

```python
def en(k,m):
    return 2.0*np.sqrt(k**2 + m**2)
```

```python
zeta = 3
M    = 3

ks = np.linspace(-3*zeta,3*zeta,2000)

fig = plt.figure(figsize=(2,3))
ax  = fig.add_subplot(1,1,1)
ax.plot(ks,[en(k,M) for k in ks],color='#6F732F',linewidth=4)
ax.plot(ks,[-en(k,M) for k in ks],color='#6F732F',linewidth=4)
ax.axis('off')
plt.tight_layout()
fig.savefig("wide_bands=3.svg")
```

```python
zeta = 3
M    = 0.5

ks = np.linspace(-3*zeta,3*zeta,2000)

fig = plt.figure(figsize=(2,3))
ax  = fig.add_subplot(1,1,1)
ax.plot(ks,[en(k,M) for k in ks],color='#6F732F',linewidth=4)
ax.plot(ks,[-en(k,M) for k in ks],color='#6F732F',linewidth=4)
ax.axis('off')
plt.tight_layout()
fig.savefig("narrow_bands.svg")
```
