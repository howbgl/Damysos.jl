# Plot "speedup"

Speedup is defined by $$speedup = \frac{t_1}{t_N} $$ where $t_i$ is the duration using $i$ processors/workers/threads.

Here I choose $t_1\rightarrow t_{100}$ since I don't have the duration of 1 yet, but this probably changes nothing. 

```python
t1 = data.duration[0]/100
plt.plot(data.nchunk[0:-6,],t1 / data.duration[0:-6])
# plt.plot(data.nchunk[0:-6,],data.nchunk[0:-6,] / 100)
```
